```python
%matplotlib inline
 
import matplotlib.pyplot as plt
import pandas as pd 
 
```

```python
df =pd.read_csv("Titanic.csv") 
df.head()
 
```

```python
df.info() 
```

## Bar

```python
 survived_counts = df.groupby("Sex")["Survived"].value_counts().unstack()
 
survived_counts.plot(kind="bar", stacked=True, figsize=(10, 10), color=["black", "red"])

 
plt.title( "Number of survivors and non-survivors by gender" )
plt.xlabel("gender" )
plt.ylabel(" Number of passengers  ")
plt.legend([" Did not survive" , "survive"])
plt.show()
```

```python
 # Count passengers in each class
pclass_counts = df["Pclass"].value_counts()

# Create the bar chart
plt.figure(figsize=(6, 4))
plt.bar(pclass_counts.index, pclass_counts.values, color="pink")

# Add labels and title
plt.title("Passenger Distribution by Class")
plt.xlabel("Class")
plt.ylabel("Number of Passengers")
plt.xticks([1, 2, 3])
plt.show()
```

## Histogram

```python
plt.figure(figsize=(8, 6))
plt.hist(df["Age"].dropna(), bins=30, color="purple", edgecolor="black", alpha=0.7)

 
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

## Bar

```python
 
survival_rates = df.groupby(["Pclass", "Sex"])["Survived"].mean().unstack()

 
survival_rates.plot(kind="bar", figsize=(8, 6), color=["orange", "blue"])

 
plt.title("Survival Rate by Class and Gender")
plt.xlabel("Class")
plt.ylabel("Survival Rate")
plt.legend(["Female", "Male"])
plt.show()
```
